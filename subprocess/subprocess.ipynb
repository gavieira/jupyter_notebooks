# Python's suprocess

The "sqlite3" module provides us with a straightforward way to run child processes from a python script. It is particularly interesting when you need the input of a shell command, for instance.

In order to use this module, we need to import it first:

```python
import subprocess
```

After the import statement, we can use the **help()** function in order to get a quick description of the module and its features:

```python
help(subprocess)
```

Documentation for this module can be found [here](https://docs.python.org/3/library/subprocess.html)
